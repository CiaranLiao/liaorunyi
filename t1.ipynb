```python
import numpy as np
import math
class Perceptor(object):
    def __init__(self,w,b):
        self.w=w
        self.b=b
    def __call__(self,x):
        return self.sigmoid(np.dot(x,self.w)+self.b)
    def sigmoid(self,z):
        return 1/(1+math.exp(-z))
```

```python
p=Perceptor([1,3,2,1,1,2],-7)
print(p([1,0,1,1,1,0]))
```

```python
import pandas as pd
data=pd.read_csv("BeijingPM20100101_20151231.csv")
data
```

```python
def Count(area):
    I=II=III=IV=C=0
    for i in area:
        if i>150:
            C+=1
            IV+=1
        elif i>75:
            C+=1
            III+=1
        elif i>35:
            C+=1
            II+=1
        elif i>0:
            C+=1
            I+=1            
    print("重度污染占比：{0:.2%},中度污染占比：{1:.2%},轻度污染占比：{2:.2%},良好占比：{3:.2%}".format(I/C,II/C,III/C,IV/C))
```

```python
Count(data['PM_Dongsi'])
Count(data['PM_Dongsihuan'])
Count(data['PM_Nongzhanguan'])
```

```python
lst=data.loc[:,['year','month','PM_Dongsi','PM_Dongsihuan', 'PM_Nongzhanguan']]
lst=lst[lst["year"].isin(['2013'])]
lst
```

```python
for i in range(1,13):
    m=lst[lst["month"].isin([i])]
    print("2013年{0}月各区PM平均值：东四{1:.2f}，东四环{2:.2f}，农展馆{3:.2f}".format(i,np.nanmean(m["PM_Dongsi"]),np.nanmean(m["PM_Dongsihuan"]),np.nanmean(m["PM_Nongzhanguan"])))
```
